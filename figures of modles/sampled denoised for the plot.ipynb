```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
from mpl_toolkits.mplot3d import Axes3D

for file in os.listdir('test_results'):
    #print(file)
    with open(os.path.join('test_results',file), 'rb') as input:
        test_res = pickle.load(input)
    noisy_signal, orginal_signal, prediction_signal = test_res
    
        
    print(f'the folder i want to know is {file}')
    print('noisy_signal.sahpe',noisy_signal.shape)
    print('orginal_signal.shape',orginal_signal.shape)
    print('prediction_signal.shape',prediction_signal.shape)
       # if you want to know more folders delete the break 
    
```

```python
# Maximal font/style settings for journal plots
LABEL_FONT_SIZE = 28   # Axis label size
TITLE_FONT_SIZE = 28   # Title size
TICK_FONT_SIZE = 24    # Tick label size
LINE_WIDTH = 4         # Plot line width

plt.rcParams.update({
    "axes.labelweight": "bold",
    "axes.titlesize": TITLE_FONT_SIZE,
    "axes.labelsize": LABEL_FONT_SIZE,
    "xtick.labelsize": TICK_FONT_SIZE,
    "ytick.labelsize": TICK_FONT_SIZE,
    "font.weight": "bold",
    "axes.linewidth": 3,
    "xtick.major.width": 2.5,
    "ytick.major.width": 2.5,
})

# Load signals
with open(os.path.join('../test_results', 'FCN_DAE_withoutresidual.pkl'), 'rb') as f:
    noisy_signal, original_signal, prediction_signal = pickle.load(f)

sample_idx = 3460
x = range(512)
noisy_ecg = noisy_signal[sample_idx, :, 0]
predicted_ecg = prediction_signal[sample_idx, :, 0]

def plot_square(signal, color, title, xlabel, ylabel):
    fig = plt.figure(figsize=(13, 10))  # even more square, high-res
    ax = fig.add_subplot(111)
    ax.plot(x, signal, color=color, linewidth=LINE_WIDTH)
    ax.set_xlabel(xlabel, fontsize=LABEL_FONT_SIZE, weight='bold', labelpad=22)
    ax.set_ylabel(ylabel, fontsize=LABEL_FONT_SIZE, weight='bold', labelpad=22)
    ax.tick_params(axis='both', which='major',
                   labelsize=TICK_FONT_SIZE, width=2.5, length=10)
    for spine in ax.spines.values():
        spine.set_linewidth(3)
    plt.tight_layout(rect=[0, 0.07, 1, 1])  # leave space at bottom for title
    plt.figtext(0.5, 0.015, title, fontsize=TITLE_FONT_SIZE + 2,
                ha='center', va='bottom', weight='bold')
    plt.show()
    plt.close()

# Plot 1: Noisy Signal
plot_square(noisy_ecg, 'red', "Noisy ECG Signal", "Sample Number", "Amplitude")

# Plot 2: Denoised Signal
plot_square(predicted_ecg, 'blue', "Denoised ECG Signal", "Sample Number", "Amplitude")
```
